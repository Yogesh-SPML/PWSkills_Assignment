Flask is a micro web framework that is written in Python. It is designed to be lightweight, flexible, and easy to use, and it is a popular choice for building web applications and APIs.

The advantages of Flask Framework are:

1. Lightweight: Flask is a micro framework, which means that it is lightweight and has a small footprint. It is easy to learn and use, and it does not require a lot of setup or configuration.

2. Flexibility: Flask is a flexible framework that allows developers to customize and extend its functionality to meet their specific needs. It provides a wide range of extensions and libraries that can be used to add features and functionality to the application.

3. Easy to Use: Flask is designed to be easy to use and understand, even for beginners. It has a simple and intuitive syntax that makes it easy to write and maintain code.

```python
#Q3. What is App routing in Flask? Why do we use app routes?
```

In Flask, application routing refers to the process of mapping URLs to view functions. A view function is a Python function that returns a response to an HTTP request. 

We use app routes in Flask to define the URLs that the application should respond to and to map those URLs to the appropriate view functions. When a user requests a URL, Flask uses the routing information to determine which view function should handle the request and generate the response.
